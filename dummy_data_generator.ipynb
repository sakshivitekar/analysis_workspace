```python
import pandas as pd
import numpy as np
from datetime import  datetime, timedelta
import random
import os
```

```python
os.makedirs("data/raw", exist_ok=True)
```

```python
num_users = 1000
```

```python
users = pd.DataFrame({
    "user_id": range(1, num_users + 1),
    "role": np.random.choice(["student", "recruiter"], num_users),
    "college_id": np.random.randint(1, 50, num_users),
    "signup_date": [
        datetime.now() - timedelta(days=random.randint(0, 60))
        for _ in range(num_users)
    ]
})
```

```python
users.to_csv("data/raw/users.csv", index = False)
```

```python
num_events = 5000
```

```python
events = pd.DataFrame({
    "event_id": range(1, num_events + 1),
    "user_id": np.random.randint(1, num_users + 1, num_events),
    "event_type": np.random.choice(
        ["signup", "profile_complete", "post", "comment", "job_view", "job_apply"],
        num_events
    ),
    "timestamp": [
        datetime.now() - timedelta(days=random.randint(0, 60))
        for _ in range(num_events)
    ]
})
```

```python
events.to_csv("data/raw/events.csv", index=False)
```

```python
print ("users.csv and events.csv created successfully!")
```
